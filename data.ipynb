```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cols = [" Frequency", "Angle of attack", "Chord length", "Velocity", " Suction side", "Pressure level"]
df = pd.read_csv("airfoil_self_noise.dat", names=cols, sep="\t")
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop("Pressure level", axis=1)
y = df["Pressure level"]
X_train, X_test, y_train, y_test =train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.info()
```

```python
for i in X_train.columns:
    sns.boxplot(data=X_train, x=i, y=y)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet, LassoCV, RidgeCV, ElasticNetCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
lassocv = LassoCV().fit(X_train, y_train)
```

```python
lassocv.alpha_
```

```python
lasso = Lasso(alpha=lassocv.alpha_)

lasso.fit(X_train, y_train)

lasso.score(X_test, y_test)
```

```python
ridgecv = RidgeCV().fit(X_train, y_train)

ridge = Ridge(alpha = ridgecv.alpha_).fit(X_train, y_train)

ridge.score(X_test, y_test)
```

```python
elasticnet = ElasticNetCV().fit(X_train, y_train)

elasticnet.alpha_
```

```python
elastic = ElasticNet(alpha=elasticnet.alpha_).fit(X_train, y_train)

elastic.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
X2 = df.drop("Pressure level", axis=1)
y2 = df["Pressure level"]

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.1, random_state=0)
```

```python
param_grid = {
    "max_depth" : [2,3,4,5],
    "min_samples_split" : [2,3,4,5,6,7],
    "min_samples_leaf" : [1,2,3,4,5]
}

random_forest = RandomForestRegressor()

grid_search = GridSearchCV(random_forest, param_grid, cv=5)
```

```python
grid_search.fit(X_train2, y_train2)
```

```python
grid_search.best_params_
```

```python
rand_forest = RandomForestRegressor(max_depth=5, min_samples_leaf=3, min_samples_split=7)
```

```python
model = rand_forest.fit(X_train2, y_train2)
```

```python
y_pred2 = model.predict(X_test2)
```

```python
r2_score(y_test2, y_pred2)
```
